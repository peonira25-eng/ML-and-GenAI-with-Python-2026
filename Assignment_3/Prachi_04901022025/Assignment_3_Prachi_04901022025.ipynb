#### **Question-2: Calculate Sum of First N Natural Numbers**

```python
# Function to calculate sum of first N natural numbers

def calculate_sum(n):

    total = 0

    for number in range(1, n + 1):
        total += number

    return total
```

```python
# Define the value of N

n = 10

# Calculate the sum

result = calculate_sum(n)

# Display the result

print(f"Sum of first {n} natural numbers = {result}")
```

#### **Question-3: Reverse a Number**

```python
# Function to reverse a given number

def reverse_number(number):

    reversed_number = 0

    while number > 0:

        digit = number % 10
        reversed_number = reversed_number * 10 + digit
        number = number // 10

    return reversed_number
```

```python
# Define the number

number = 12345

# Reverse the number

result = reverse_number(number)

# Display the result

print(f"Original Number = {number}")
print(f"Reversed Number = {result}")
```

#### **Question-4: Count Digits in a Number**

```python
# Function to count digits in a number

def count_digits(number):

    digit_count = 0

    while number > 0:

        digit_count += 1
        number = number // 10

    return digit_count
```

```python
# Define the number

number = 123456

# Count the digits

result = count_digits(number)

# Display the result

print(f"Number of Digits = {result}")
```

#### **Question-5: Check Palindrome Number**

```python
# Function to check whether a number is palindrome

def check_palindrome(number):

    original_number = number
    reversed_number = 0

    while number > 0:

        digit = number % 10
        reversed_number = reversed_number * 10 + digit
        number = number // 10

    return original_number == reversed_number
```

```python
# Define the number

number = 121

# Check palindrome status

result = check_palindrome(number)

# Display the result

if result:
    print(f"{number} is a Palindrome Number")
else:
    print(f"{number} is not a Palindrome Number")
```

#### **Question-6: Generate Fibonacci Series**

```python
# Function to generate Fibonacci series

def generate_fibonacci(term_count):

    fibonacci_series = []

    first_term = 0
    second_term = 1

    for _ in range(term_count):

        fibonacci_series.append(first_term)

        next_term = first_term + second_term
        first_term = second_term
        second_term = next_term

    return fibonacci_series
```

```python
# Define the number of terms

term_count = 10

# Generate Fibonacci series

result = generate_fibonacci(term_count)

# Display the series

print("Fibonacci Series:")
print(result)
```

#### **Question-7: Calculator Using Functions**

```python
# Functions for basic arithmetic operations

def add_numbers(first_number, second_number):
    return first_number + second_number

def subtract_numbers(first_number, second_number):
    return first_number - second_number

def multiply_numbers(first_number, second_number):
    return first_number * second_number

def divide_numbers(first_number, second_number):

    if second_number == 0:
        return "Division by zero is not allowed"

    return first_number / second_number
```

```python
# Sample values

num1 = 25
num2 = 5

# Select operation

operation = "Division"

# Perform calculation

if operation == "Addition":
    result = add_numbers(num1, num2)

elif operation == "Subtraction":
    result = subtract_numbers(num1, num2)

elif operation == "Multiplication":
    result = multiply_numbers(num1, num2)

elif operation == "Division":
    result = divide_numbers(num1, num2)

else:
    result = "Invalid Operation"

# Display result

print(f"Operation: {operation}")
print(f"First Number: {num1}")
print(f"Second Number: {num2}")
print(f"Result: {result}")
```

#### **Question-10: Handle Division by Zero Using Exception Handling**

```python
# Function to perform safe division

def safe_division(dividend, divisor):

    try:

        result = dividend / divisor
        return result

    except ZeroDivisionError:

        return "Division by zero is not allowed"
```

```python
# Sample values

number1 = 20
number2 = 0

# Perform division

result = safe_division(number1, number2)

# Display result

print(f"Result: {result}")
```
