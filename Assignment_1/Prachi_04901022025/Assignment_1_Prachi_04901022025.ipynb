#### **Question-1: Find Area of Rectangle**

```python
# Function to calculate the area of a rectangle

def calculate_area(length, width):
    area = length * width
    return area
```

```python
# Rectangle dimensions

length = 12
width = 8

# Compute area using the defined function

result = calculate_area(length, width)

# Display the result

print(f"Area of Rectangle = {result}")
```

#### **Question-2: Calculate Simple Interest**

```python
# Function to calculate Simple Interest

def calculate_simple_interest(principal, rate, time):
    simple_interest = (principal * rate * time) / 100
    return simple_interest
```

```python
# Given financial details

principal = 10000
rate = 5
time = 2

# Calculate Simple Interest

result = calculate_simple_interest(principal, rate, time)

# Display the result

print(f"Simple Interest = {result}")
```

#### **Question-3: Convert Temperature from Celsius to Fahrenheit**

```python
# Function to convert Celsius to Fahrenheit

def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9 / 5) + 32
    return fahrenheit
```

```python
# Temperature in Celsius

celsius = 37

# Perform temperature conversion

result = celsius_to_fahrenheit(celsius)

# Display the converted temperature

print(f"Temperature in Fahrenheit = {result}")
```

#### **Question-4: Calculate the Average of Three Numbers**

```python
# Function to calculate the average of three numbers

def calculate_average(num1, num2, num3):
    average = (num1 + num2 + num3) / 3
    return average
```

```python
# Sample values

num1 = 25
num2 = 40
num3 = 35

# Compute the average value

result = calculate_average(num1, num2, num3)

# Display the result

print(f"Average = {result}")
```

#### **Question-5: Find the Square and Cube of a Number**

```python
# Function to calculate square and cube of a number

def calculate_power_values(number):
    square = number ** 2
    cube = number ** 3
    return square, cube
```

```python
# Sample value

number = 5

# Compute square and cube

square, cube = calculate_power_values(number)

# Display the results

print(f"Square = {square}")
print(f"Cube = {cube}")
```

#### **Question-6: Swap Two Numbers Without Using a Third Variable**

```python
# Function to swap two numbers without using an additional variable

def swap_numbers(first_number, second_number):
    first_number, second_number = second_number, first_number
    return first_number, second_number
```

```python
# Sample values

num1 = 15
num2 = 25

# Display values before swapping

print("Before Swapping:")
print(f"First Number = {num1}")
print(f"Second Number = {num2}")

# Perform swapping operation

num1, num2 = swap_numbers(num1, num2)

# Display values after swapping

print("\nAfter Swapping:")
print(f"First Number = {num1}")
print(f"Second Number = {num2}")
```
