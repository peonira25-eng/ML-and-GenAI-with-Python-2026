#### **Question-1: Find the Sum of the First 10 Natural Numbers**

```python
# Function to calculate the sum of natural numbers

def calculate_sum(limit):
    total = sum(range(1, limit + 1))
    return total
```

```python
# Define the upper limit

number_limit = 10

# Calculate the sum

result = calculate_sum(number_limit)

# Display the result

print(f"Sum of First 10 Natural Numbers = {result}")
```

#### **Question-2: Find the Factorial of a Number**

```python
# Function to calculate the factorial of a number

def calculate_factorial(number):
    factorial = 1

    for value in range(1, number + 1):
        factorial *= value

    return factorial
```

```python
# Define the number

number = 5

# Calculate factorial

result = calculate_factorial(number)

# Display the result

print(f"Factorial of {number} = {result}")
```

#### **Question-3: Print Fibonacci Series**

```python
# Function to generate Fibonacci series

def generate_fibonacci(term_count):
    fibonacci_series = []

    first_term = 0
    second_term = 1

    for _ in range(term_count):
        fibonacci_series.append(first_term)
        first_term, second_term = second_term, first_term + second_term

    return fibonacci_series
```

```python
# Define the number of terms

term_count = 10

# Generate Fibonacci series

result = generate_fibonacci(term_count)

# Display the series

print("Fibonacci Series:")
print(result)
```

#### **Question-4: Find the Largest Among Three Numbers**

```python
# Function to determine the largest number

def find_largest(first_number, second_number, third_number):

    if first_number >= second_number and first_number >= third_number:
        return first_number

    elif second_number >= first_number and second_number >= third_number:
        return second_number

    else:
        return third_number
```

```python
# Define three numbers

num1 = 45
num2 = 78
num3 = 62

# Find the largest number

result = find_largest(num1, num2, num3)

# Display the result

print(f"Largest Number = {result}")
```

#### **Question-5: Student Result System**

```python
# Function to generate student performance report

def generate_result(student_name, roll_number, marks):

    total_marks = sum(marks)
    percentage = total_marks / len(marks)

    if percentage >= 90:
        grade = "A+"
    elif percentage >= 80:
        grade = "A"
    elif percentage >= 70:
        grade = "B"
    elif percentage >= 60:
        grade = "C"
    else:
        grade = "D"

    result = {
        "Name": student_name,
        "Roll Number": roll_number,
        "Total Marks": total_marks,
        "Percentage": percentage,
        "Grade": grade
    }

    return result
```

```python
# Sample student record

student_name = "Prachi"
roll_number = "049"

# Marks obtained in Python, Machine Learning and Generative AI

marks = [90, 95, 88]

# Generate student report

student_report = generate_result(
    student_name,
    roll_number,
    marks
)

# Display report

print("----- Student Report -----")

for key, value in student_report.items():

    if key == "Percentage":
        print(f"{key}: {value:.2f}%")
    else:
        print(f"{key}: {value}")
```
