## Q1. Loan Approval Prediction
**A bank wants to predict whether a loan application will be approved or rejected.**

**Is this a classification or regression problem?**

This is a **Classification** problem because the output is a category — either 'Approved' or 'Rejected'. There are only two possible outcomes, not a continuous number.

**Which algorithm would be most suitable?**

**Decision Tree** is most suitable because:
- It works well with both numerical and categorical data (age, income, credit score, etc.)
- It gives clear rules for decision making (easy to explain to bank managers)
- KNN can also work but is slower on large datasets
- Linear Regression is not suitable here because it predicts numbers, not categories

## Q3. How Decision Tree Makes a Prediction
**A Decision Tree predicts that a customer will purchase a product. Explain how the model reaches this prediction.**

A Decision Tree works like a flowchart — it asks a series of yes/no questions at each node and follows a path based on the answers.

**Example — Customer Purchase Prediction:**

1. **Root Node:** Is the customer's age > 25? → Yes
2. **Next Node:** Is monthly income > ₹30,000? → Yes  
3. **Next Node:** Has the customer visited the website more than 3 times? → Yes
4. **Leaf Node:** → **Prediction: Will Purchase**

The tree was trained on past customer data, so it learned which conditions lead to a purchase. During prediction, it simply follows the learned rules from root to leaf node.

## Q4. KNN as Supervised Learning
**Why is KNN considered a supervised learning algorithm? Can KNN perform classification if labels are not available?**

**Why KNN is supervised learning:**

KNN is supervised because it requires **labeled training data** to make predictions. During training, it stores all data points along with their labels. When a new data point comes, it finds the K nearest neighbors and uses their labels to predict the output.

Without labels, KNN cannot learn which class a data point belongs to.

**Can KNN work without labels?**

No, KNN **cannot** perform classification without labels. Labels are essential because KNN uses majority voting among neighbors — if there are no labels, there is nothing to vote on. Without labels, we would need an unsupervised algorithm like K-Means Clustering instead.

## Q5. KNN Prediction with K=5
**K = 5, Neighbors: Yes, Yes, No, Yes, No — What will be the prediction?**

**Prediction: Yes**

**Reason:**

KNN uses **majority voting** among K nearest neighbors.

| Neighbor | Vote |
|---|---|
| 1st | Yes |
| 2nd | Yes |
| 3rd | No |
| 4th | Yes |
| 5th | No |

- **Yes count: 3**
- **No count: 2**

Since **Yes** has the majority (3 out of 5), the final prediction is **Yes**.

## Q6. Linear Regression Equation
**Explain: y = mx + b**

The equation **y = mx + b** is the equation of a straight line used in Linear Regression.

| Symbol | Name | Meaning |
|---|---|---|
| **x** | Feature/Input | The independent variable (what we know) |
| **y** | Label/Output | The dependent variable (what we predict) |
| **m** | Slope | How much y changes when x increases by 1 |
| **b** | Intercept | The value of y when x = 0 |

**Example — Predicting House Price:**
- x = size of house (sq ft)
- y = price of house (₹)
- m = price increase per sq ft
- b = base price even if size is 0

If m = 5000 and b = 100000, then a 200 sq ft house costs: y = 5000×200 + 100000 = ₹11,00,000

## Q7. Correlation
**What is correlation? What do +1, 0, and -1 indicate?**

**Correlation** measures the strength and direction of the relationship between two variables. It tells us how one variable changes when another changes.

| Value | Meaning | Example |
|---|---|---|
| **+1** | Perfect positive correlation | As study hours increase, marks also increase |
| **0** | No correlation | Shoe size and intelligence — no relation |
| **-1** | Perfect negative correlation | As temperature increases, sweater sales decrease |

- Values close to +1 mean strong positive relationship
- Values close to -1 mean strong negative relationship
- Values close to 0 mean weak or no relationship

## Q8. Disease Detection Error
**Actual: Disease Present → Predicted: Healthy. What type of error is this and why is it dangerous?**

**Type of Error: False Negative**

A False Negative occurs when the model predicts **Negative (Healthy)** but the actual condition is **Positive (Disease Present)**.

**Why is it dangerous?**

- The patient actually has the disease but the model says they are healthy
- The patient will not receive treatment because they think they are fine
- The disease can spread or become worse without treatment
- In serious diseases like cancer, a False Negative can even lead to death
- This is considered the most dangerous type of error in medical diagnosis

## Q9. Confusion Matrix vs Correlation Matrix
**What is the purpose of a Confusion Matrix? How is it different from a Correlation Matrix?**

**Confusion Matrix:**
- Used to evaluate the performance of a **classification model**
- Shows how many predictions were correct and incorrect
- Contains: True Positives, True Negatives, False Positives, False Negatives

| | Predicted: Yes | Predicted: No |
|---|---|---|
| **Actual: Yes** | True Positive | False Negative |
| **Actual: No** | False Positive | True Negative |

**Correlation Matrix:**
- Used to show the **relationship between features** in a dataset
- Values range from -1 to +1
- Helps identify which features are related to each other

**Key Difference:**

| | Confusion Matrix | Correlation Matrix |
|---|---|---|
| Purpose | Evaluate model accuracy | Find feature relationships |
| Used for | Classification results | Data analysis |
| Values | Counts (0, 1, 2...) | -1 to +1 |

## Q10. Algorithm Comparison
**Compare: Decision Tree, KNN, SVM, Linear Regression, Neural Network**

| Algorithm | Main Idea |
|---|---|
| **Decision Tree** | Splits data using yes/no questions like a flowchart to reach a decision at the leaf node |
| **KNN** | Finds K nearest neighbors in training data and uses majority vote for classification |
| **SVM** | Finds the best boundary (hyperplane) that separates two classes with maximum margin |
| **Linear Regression** | Fits a straight line through data points to predict continuous numerical values |
| **Neural Network** | Mimics the human brain using layers of connected nodes to learn complex patterns |

**When to use which:**
- Decision Tree → When you need explainable results
- KNN → When dataset is small and simple
- SVM → When classes are clearly separable
- Linear Regression → When predicting a number
- Neural Network → When data is very complex (images, speech)
